```python
from skimage.io import imread, imsave, imshow
from skimage import img_as_float, img_as_ubyte
import numpy as np

# reading image to img variable 
img = imread('https://stepik.org/media/attachments/lesson/58402/tiger-low-contrast.png')

# converting image colors to range 0...1
img = img.astype('float')

# getting values and bins of histogram
values, bins = np.histogram(img.ravel(), bins=range(257))

# initializing amout of pixels we want to drop
k = round(img.size * 0.05)

# sorting image
sorted_img = sorted(img.ravel())

# getting min and max value
min_b = sorted_img[k + 1]
max_b = sorted_img[-(k+1)]

# evaluating new correlated image
img_correlated = (img - min_b) * (255 / (max_b - min_b))

# cutting img to range 0...255
img_correlated = np.clip(img_correlated, 0, 255)

# coverting image to uint8
img = img_correlated.astype('uint8')

# saving correlated image
imsave('out_img.png', img)
```

```python
%matplotlib inline
imshow(img)
```

```python
# reading image to img variable 
img = imread('https://stepik.org/media/attachments/lesson/60609/tiger-color.png')
```

```python
# converting colors to 0..1 range
img_f = img_as_float(img)
```

```python
# making a copy of our image
img_copy = img_f.copy()

# converting rgb colors to yuv
img_copy[:, :, 0] = img_f[:, :, 0] * 0.2126 + img_f[:, :, 1] * 0.7152 + img_f[:, :, 2] * 0.0722
img_copy[:, :, 1] = - img_f[:, :, 0] * 0.0999 - img_f[:, :, 1] * 0.3360 + img_f[:, :, 2] * 0.4360
img_copy[:, :, 2] = img_f[:, :, 0] * 0.6150 - img_f[:, :, 1] * 0.5586 - img_f[:, :, 2] * 0.0563
```

```python
# initializing amout of pixels we want to drop
k = round(img_copy[:, :, 0].size * 0.05)

# sorting image
sorted_img = sorted(img_copy[:, :, 0].ravel())

# getting min and max value
min_b = sorted_img[k + 1]
max_b = sorted_img[-(k+1)]

# evaluating new correlated image
img_correlated_y = (img_copy[:, :, 0] - min_b) / (max_b - min_b)
```

```python
# clipping values from 0 to 1
img_correlated_y = np.clip(img_correlated_y, 0, 1)
img_copy[:, :, 0] = img_correlated_y
```

```python
# converting image back to rgb
img_f[:, :, 0] = img_copy[:, :, 0] + 1.2803 * img_copy[:, :, 2]
img_f[:, :, 1] = img_copy[:, :, 0] - 0.2148 * img_copy[:, :, 1]  - 0.3805 * img_copy[:, :, 2]
img_f[:, :, 2] = img_copy[:, :, 0] + 2.1279 * img_copy[:, :, 1]
```

```python
# clipping values from 0 to 1
img_f = np.clip(img_f, 0, 1)

# converting colors to 0..255
img_f = img_as_ubyte(img_f)
```

```python
# saving image to file
imsave('out_img.png', img_f)
```

```python
# showing image
imshow(img_f)
```

```python
# reading image to img variable 
img = imread('https://stepik.org/media/attachments/lesson/60610/railroad.png')

# converting colors to range 0...1
img = img_as_float(img)
```

```python
# getting cannels averages
r_av = img[:, :, 0].mean()
g_av = img[:, :, 1].mean()
b_av = img[:, :, 2].mean()

# getting average
avg = (r_av + g_av + b_av) / 3

# getting coefficients
r_w = r_av / avg
g_w = g_av / avg
b_w = b_av / avg
```

```python
# dividing channels by coefficients
img[:, :, 0] /= r_w
img[:, :, 1] /= g_w
img[:, :, 2] /= b_w
```

```python
# clipping image from 0 to 1
img = np.clip(img, 0, 1)
```

```python
# saving image to file
imsave('out_img.png', img)
```
