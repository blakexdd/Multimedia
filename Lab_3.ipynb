```python
#FILE NAME: Lab_3
#PROGRAMMER: VG6
#DATE: 14.03.2020
#Purpose: To learn metrics

# importing libraries
import numpy as np
from skimage.io import imread, imshow
from Convertion_lib import decode_yuv, encode_rgb, decrease_brights_resolution

# Evaluating entropy function
# Arguments:
#   - image we want to calculate entropy of
#     and functional algoritm
#       img : np.array, func1 : def, func2 : def
# Returns:
#   - evaluated entropy
#      (float)
def entropy(img : np.array, func1, func2 = None):
    # getting initial histogram
    values, bins = np.histogram(img.ravel(), bins=range(257))
    
    # getting probabilities out of values
    probs = values / values.sum()
    
    # getting entropy
    H = -probs * np.log2(probs)
    
    if func2 == None:
        # converting image
        converted_img = func1(img, 4)
    else:
        temp_img = func1(img)
        converted_img = func2(temp_img)
        
    # getting converted histogram
    values_conv, bins = np.histogram(converted_img.ravel(), bins=range(257))
    
    print(values_conv)
    # getting convetred probabilities
    probs_conv = values_conv / values_conv.sum()
    
    # getting converted entropy
    H_conv = -probs_conv * np.log2(probs_conv)
    
    return H.sum() , H_conv.sum()
    
    
```
