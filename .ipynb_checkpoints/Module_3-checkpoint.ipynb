```python
# Stepik Module 3 tasks implementing by Vladimir Gololobov
# Purpose: To learn how to implement color and bright correlation

# importing modules
from skimage.io import imread, imsave, imshow
from skimage import img_as_float
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
```

```python
# reading image to img variable
img = imread('https://stepik.org/media/attachments/lesson/58402/tiger-low-contrast.png')

# showing image
imshow(img)
```

```python
# painting brights histogram and see, that it needs correlation
values, bins, _ = plt.hist(img.ravel(), bins=range(257))
```

```python
# getting max and min values of bright 
b_max = img.max()
b_min = img.min()

# evaluating new correlated image
img_correlated = (img - b_min) * (255 / (b_max - b_min))

# converting colors to range 0...255
img_correlated = img_correlated.astype('uint8')

# painting new brights histogram and see, what 
# we have brights for almost all values from 0 to 255
values_c, bins_c, _ = plt.hist(img_correlated.ravel(), bins=range(257))
```

```python
#showing correlated image 
imshow(img_correlated)
```

```python
# saving image to file 
imsave('out_img.png', img_correlated)
```

```python
# reading image to img variable 
img = imread('https://stepik.org/media/attachments/lesson/58402/tiger-low-contrast.png')

# showing image
imshow(img)
```

```python
# converting image colors to range 0...1
img = img.astype('float')


# getting values and bins of histogram
values, bins = np.histogram(img.ravel(), bins=range(257))
```

```python
# getting number of pixels with not 0 value
brights_number = len(values[values != 0])

# initializing amout of pixels we want to drop
k = round(img.size * 0.05)

# sorting image
sorted_img = sorted(img.ravel())

# getting min and max value
min_b = sorted_img[k + 1]
max_b = sorted_img[-(k+1)]
```

```python
# printing max and min values
print(min_b, max_b)
```

```python
# evaluating new correlated image
img_correlated = (img - min_b) * (255 / (max_b - min_b))

# cutting img to range 0...255
img_correlated = np.clip(img_correlated, 0, 255)

# coverting image to uint8
img = img_correlated.astype('uint8')
```

```python
# showing correlated image
imshow(img)
```

```python
# saving correlated image
imsave('out_img.png', img)
```
